```python
# Lets Start with importing the  matplot library
import matplotlib.pyplot as plt
# To create dataframe and array lets import pandas and numpy
import pandas as pd
import numpy as np
```

```python
# we creating the sample databy using numpy array
a = np.linspace(10,50,5)
print(a)
b = np.array([100,200,300,400,500])  # or b=np.linspace(100,500,5)
print(b)
```

```python
# Now Ploting
plt.figure(figsize=(8,6))
plt.plot(a,b,label ='RPM RATE')
plt.title('Line Plot')
plt.xlabel('Frq')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

# Histogram
* Histogram take only one varible value
* To construct a histogram, the first step is to "bin" the range of valuesthat is, divide the entire range of values into a series of intervalsand then count how many values fall into each interval
*  A histogram is used for continuous data
*  The height of each bar in a histogram indicates the frequency of data points within each bin.
*  

```python
# create random data
k = np.random.randint(1,50,100)
plt.hist(k,bins=10, edgecolor='black')
plt.grid(True)
plt.title('Scatter PLot')
plt.ylabel('Freq')
plt.xlabel('Value')
plt.show()
```

```python
# Without egde color and 20 bins
# create random data
k = np.random.randint(1,50,100)
plt.hist(k,bins=20)
plt.grid(True)
plt.title('Scatter PLot')
plt.ylabel('Freq')
plt.xlabel('Value')
plt.show()
```

# Bar Plot
* A bar chart is a type of chart that presents categorical data with
rectangular bars. The lengths of the bars are proportional to the
values they represent. Bar charts are particularly useful for comparing the values of different categories or groups.
* Bar charts can be used to analyze trends over a period by representing data in different time intervals.
* When dealing with categorical data, such as types of products, regions, or groups, bar charts are a suitable choice.

* Bar charts provide a straightforward and clear visual representation of data, making them accessible for a wide audience.

### Vertical barplot
### Horizontal bar plot

```python
# Creating the categorical data
name = ['a','b','c','d','e','f']
age = [25,26,14,36,45,50]
```

```python
plt.bar(name,age,color='skyblue')
plt.title(' Vertical Bar pLot')
plt.xlabel('Name')
plt.ylabel('age')
plt.show()

# Both variables required of same size 
```

```python
# Horizontal bar plot
plt.barh(name,age,color='orange')
plt.title(' Vertical Bar pLot')
plt.xlabel('Name')
plt.ylabel('age')
plt.show()
```

# Box PLot
* Using the bar plot we can detect the outliers which are present in the  data
* simple to apply
* less time consuming than other methods

```python
# Sample data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Plot

plt.boxplot(data, patch_artist=True, labels=['x1', 'x2', 'x3'])
plt.title('Box Plot')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

 # Subplots
 * subplot like as picture in picture

```python

plt.subplot(1, 3, 1)  
plt.plot([1, 2, 3,4,5], [10,20,25,40,45],marker='o',
         color='green',linestyle=':')
plt.title('A')

plt.subplot(1, 3, 2)  
plt.plot([1, 2, 3,4], [12,14,20,25],color='blue'
         ,marker='p',linestyle='-.')
plt.title("B")

plt.subplot(1,3,3)  
plt.plot([1, 2, 3,4,5], [11,20,33,40,55],marker='D'
         ,color='orange',linestyle='--')
plt.title('C')

plt.tight_layout()  

plt.show()
```

```python
plt.subplot(3, 1, 1)  # when we change the sequence of plot number then positions changed
plt.plot([1, 2, 3,4,5], [10,20,25,40,45],marker='o',
         color='green',linestyle=':')
plt.title('A')

plt.subplot(3, 1, 2)  
plt.plot([1, 2, 3,4], [12,14,20,25],color='blue'
         ,marker='p',linestyle='-.')
plt.grid(axis='y') # with  grid on y axis
plt.title("B")

plt.subplot(3,1,3)  
plt.plot([1, 2, 3,4,5], [11,20,33,40,55],marker='D'
         ,color='orange',linestyle='--')
plt.title('C')


plt.tight_layout()  
plt.grid(axis='x') # with grid on x axis
plt.show()
```

```python
# Modifcation of Grid
plt.plot([1, 2, 3,4,5], [11,20,33,40,55],marker='D'
         ,color='orange',linestyle='--')
plt.title('After Modification')

plt.grid(linestyle='-.',
         color='green',
         linewidth=1
        )
plt.show()
```

```python
import seaborn as sns
```

# Heatmap
* For to detect the correlation between variuos varibles

```python
# create a random data
df = np.random.rand(10,10) # matrix of 10 by 10
df
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df,
            annot=True,  # for value detect
            cmap='viridis' # color
           )
plt.title('Heatmap')
plt.show()
```

# Pairplots

```python
# now we using built in  data from seabrn 
# data set name = iris
data = sns.load_dataset('iris')
data
```

```python
sns.pairplot(data,hue='species')
plt.suptitle('Pair Plot of Iris Dataset',
             y=1.02)

plt.show()
```

# Voilin Plot
* structure like an voilin

```python
#using built in data
data =sns.load_dataset('tips')

print(data)
df = pd.DataFrame(data, columns=['day', 'total_bill'])  # Replace with your actual column names
df
```

```python

plt.figure(figsize=(8, 6))
sns.violinplot(x='day', y='total_bill', data=df, palette='pastel')
plt.title('Violin Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.grid(True)
plt.show()

```

# Distribution Plot(KDE)


```python
plt.figure(figsize=(12,8))
sns.kdeplot(data,fill=True)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('KDE PLot')
plt.grid(linestyle='-.',color='red')
plt.show()
```

```python
# we can take a single variable
plt.figure(figsize=(10,6))
sns.kdeplot(df,fill=True)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('KDE PLot')
plt.grid(linestyle='-.',color='red')
plt.show()
```

```python
# for random data
rnd = np.random.randn(1000)


sns.kdeplot(rnd, fill=True)
plt.title('KDE Plot of Random Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

# Joint plots

```python
o=[100,200,300,400,450,500,600,650,700,800]
p=[12,14,16,18,18,22,26,24,28,30]
```

```python
plt.figure(figsize=(10,8))
sns.jointplot(x=o,y=p,kind='scatter')
plt.suptitle('Joint Plot of x and y', y=1.09)
plt.grid(True)
plt.show()
```

```python
# Using large random data
a = np.random.randn(1000)
b = (2 * a ) + np.random.randn(1000)


sns.jointplot(x=a, y=b, kind='scatter')
plt.suptitle('Joint Plot of x and y', y=1.02)
plt.show()
```
